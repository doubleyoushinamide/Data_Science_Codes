```python
import numpy as np
import pandas as pd
import matplotlib .pyplot as plt
%matplotlib inline
```

```python
df = pd.read_csv('joint.csv')
```

```python
df
```

```python
df['Biomass'] = df['Biomass'].astype(str) #This converts the dates to strings: useful as index
```

```python
df = df.set_index('Biomass') # This line finally sets df['Biomass'] as the index column
```

```python
plt.rcParams['font.family'] = ''
# By using subplots, fragments of each plots are displayed for each columns
df.plot(kind='box', subplots=True, sharex=False, use_index=True, grid=True, style=['classic'], ylabel='Biomass Crop Usage accross the UK (tonnes)', sharey=False, rot=(0), figsize=(15,10))
plt.savefig('.svg')
```

```python
# Rolling window for each columns at rolling window = 2 years
df.rolling(2).mean().plot(kind='line', subplots=True, grid=True, sharex=True, use_index=True, figsize=(15,10), xlabel=('Moving average by 5 years between 1992-2050'))

plt.savefig('Rolling_window.svg') #Save a clear plot with .svg
```
