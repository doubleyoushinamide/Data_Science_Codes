```python
import numpy as np
import pandas as pd
import matplotlib .pyplot as plt
%matplotlib inline
```

```python
df = pd.read_csv('cases_projection.csv')
```

```python
df
```

```python
df = df.set_index('Biomass') # This line finally sets df['Biomass'] as the index column
```

```python
df['Ex_Biomass'] = df['Biomass'].astype(str)
```

```python
plt.rcParams['font.family'] = ''
# By using subplots, fragments of each plots are displayed for each columns
df.plot(kind='box', subplots=True, sharex=True, use_index=True, grid=True, style=['classic'], xlabel='Biomass 3-fold prediction in the UK from from 2022-2050', sharey=False, rot=(90), figsize=(15,10))
plt.savefig('cases_plot.svg', bbox_inches='tight')
```

```python
# Rolling window for each columns at rolling window = 2 years
df.rolling(2).mean().plot(kind='line', subplots=True, grid=True, sharex=True, use_index=True, figsize=(15,10), xlabel=('Moving average of biomass export: A projection from 2009-2050'))

plt.savefig('export_Rolling_window_projection.svg') #Save a clear plot with .svg
```
