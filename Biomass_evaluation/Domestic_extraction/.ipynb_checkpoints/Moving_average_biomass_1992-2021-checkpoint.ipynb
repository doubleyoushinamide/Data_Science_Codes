```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df = pd.read_csv('biomass_222.csv')
```

```python
df.head(2)
```

```python
df = df.set_index('Biomass') # This line finally sets df['Biomass'] as the index column
```

```python
df['Biomass'] = df['Biomass'].astype(str) #This converts the dates to strings: useful as index
```

```python
df
```

```python
plt.rcParams['font.family'] = ''
# By using subplots, fragments of each plots are displayed for each columns
df.plot(kind='box', subplots=True, sharex=True, use_index=True, grid=False, style=['classic'], widths=0.5, fontsize=12, ylabel='Domestic extraction (1000 metric tonnes)', sharey=True, rot=(90), figsize=(15,10))

plt.show()
plt.savefig('Domestic_Extraction.svg', bbox_inches='tight')
```

```python
# Rolling window for each columns at rolling window = 2 years
df.rolling(30).mean().plot(kind='line', fontsize=12, subplots=True, linewidth=3, grid=False, sharex=True, use_index=True, figsize=(15,10), xlabel=('Moving average by 5 years between 1992-2050'))

plt.savefig('Rolling_window.svg') #Save a clear plot with .svg
```
