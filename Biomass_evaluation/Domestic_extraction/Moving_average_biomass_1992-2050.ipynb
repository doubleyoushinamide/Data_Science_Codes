```python
import numpy as np
import pandas as pd
import matplotlib .pyplot as plt
%matplotlib inline
```

```python
df = pd.read_csv('biomass_1992-202233.csv')
```

```python
df.head(2)
```

```python
df
```

```python
df = df.set_index('Biomass') # This line finally sets df['Biomass'] as the index column
```

```python
df['Biomass'] = df['Biomass'].astype(str) #This converts the dates to strings: useful as index
```

```python
df = df.dropna(axis=1)
```

```python
df #Check if index had
```

```python
# By using subplots, fragments of each plots are displayed for each columns
plt.rcParams['font.family'] = ''
df.plot(kind='box', subplots=True, grid=True, sharex=True, use_index=True, style=['classic'], ylabel='Domestic extraction (1000 metric tonnes)', sharey=True, rot=(15), figsize=(15,10))
plt.savefig('Domestic_Extraction_2050.svg')
```

```python
# Rolling window for each columns at rolling window = 2 years
df.rolling(2).mean().plot(kind='line', subplots=True, sharex=True, grid=True, use_index=True, figsize=(15,10), xlabel=('Moving average by 5 years between 1992-2021'))

plt.savefig('Rolling_window_2050.svg') #Save a clear plot with .svg
```
