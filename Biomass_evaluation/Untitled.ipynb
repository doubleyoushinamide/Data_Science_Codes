```python
# Assuming you have loaded your dataset into biomass_df

window_size = 5  # Adjust the window size as needed
years_before_2050 = list(range(1992, 2051))  # Years up to 2050
```

```python
# Create a new DataFrame to store the moving averages
moving_avg_df = df.copy()
```

```python
# Iterate through each row
for index, row in df.iterrows():
    values = row[1:]  # Exclude the first column 'biomass source'
    moving_avgs = np.convolve(values, np.ones(window_size)/window_size, mode='valid')
    
    # Calculate the moving averages up to the year 2050
    moving_avg_up_to_2050 = np.concatenate((np.zeros(len(years_before_2050) - len(moving_avgs)), moving_avgs))
    moving_avg_df.loc[index, years_before_2050] = moving_avg_up_to_2050
```
