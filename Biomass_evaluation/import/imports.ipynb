```python
import numpy as np
import pandas as pd
import matplotlib .pyplot as plt
%matplotlib inline
```

```python
df = pd.read_csv('import_projection.csv')
```

```python
df.head(2)
```

```python
df['Im_Biomass'] = df['Im_Biomass'].astype(str)
```

```python
df = df.set_index('Im_Biomass') # This line finally sets df['Biomass'] as the index column
```

```python
plt.rcParams['font.family'] = ''
# By using subplots, fragments of each plots are displayed for each columns
#df.plot(kind='box', subplots=True, sharex=False, use_index=True, grid=True, style=['classic'], ylabel='Biomass importation in the UK (tonnes)', sharey=True, rot=(0), figsize=(15,10))
#-------------
boxprops=dict(linewidth=2)
capprops=dict(linewidth=2)
whiskerprops=dict(linewidth=2)
# By using subplots, fragments of each plots are displayed for each columns
df.plot(kind='box', boxprops=boxprops, capprops=capprops, 
        whiskerprops=whiskerprops, subplots=True, sharex=False, use_index=True, 
        grid=False, style=['classic'], widths=0.5, fontsize=12, sharey=True, rot=(90), 
        figsize=(12,8), ylabel='Biomass importation in the UK: A projection into 2050 (tonnes)')

#plt.show()

plt.savefig('import_projection.svg', bbox_inches='tight')
```

```python
# Rolling window for each columns at rolling window = 2 years
#df.rolling(2).mean().plot(kind='line', subplots=True, grid=True, sharex=True, use_index=True, figsize=(15,10), xlabel=('Moving average of biomass import'))

# Rolling window for each columns at rolling window = 2 years
df.rolling(2).mean().plot(kind='line', fontsize=14, subplots=True, 
                           linewidth=3, grid=False, sharex=True, use_index=True, 
                           figsize=(12,8),
                           xlabel='Moving average of biomass import: A projection into 2050')
plt.savefig('import_Rolling_window_projection.svg', bbox_inches='tight') #Save a clear plot with .svg
```
