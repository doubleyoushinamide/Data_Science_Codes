```python
import numpy as np
import pandas as pd
import matplotlib .pyplot as plt
%matplotlib inline
```

```python
df = pd.read_csv('import.csv')
```

```python
df.head(2)
```

```python
df['Biomass'] = df['Biomass'].astype(str)
```

```python
df = df.set_index('Biomass') # This line finally sets df['Biomass'] as the index column
```

```python
plt.rcParams['font.family'] = ''
# By using subplots, fragments of each plots are displayed for each columns
df.plot(kind='box', subplots=True, sharex=False, use_index=True, grid=True, style=['classic'], ylabel='Biomass importation in the UK (tonnes)', sharey=True, rot=(0), figsize=(15,10))
plt.savefig('import.svg')
```

```python
# Rolling window for each columns at rolling window = 2 years
df.rolling(2).mean().plot(kind='line', subplots=True, grid=True, sharex=True, use_index=True, figsize=(15,10), xlabel=('Moving average of biomass import'))

plt.savefig('import_Rolling_window.svg') #Save a clear plot with .svg
```
